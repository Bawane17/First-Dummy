```python
df_final.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
Ratings = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(
                                                                                        columns = {0:'Rating Count'})
```

```python
Ratings
```

## Observation

1. When Rating is between 4.5 to 4.9 ----> Excellent
2. When Rating is between 4.0 to 4.4 ----> Very Good
3. When Rating is between 3.5 to 3.9 ----> Good
4. When Rating is between 2.5 to 3.4 ----> Average
5. When Rating is between 1.8 to 2.4 ----> Poor
6. When Rating is 0.0 ----> Not Rated

```python
plt.subplots(figsize=(10,5))
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = Ratings)
```

```python
plt.subplots(figsize=(15,5))
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = Ratings, palette = 
            ['black', 'red', 'orange', 'yellow', 'green','green'])
## For visibility white color represent with black color
```

## Observation:

1. Not Rated counts are very high
2. Maximum number of Ratings are between 2.6 to 3.9

```python
## Count plot
plt.subplots(figsize = (13,5))
sns.countplot(x = 'Rating color', data = Ratings, palette = ['black', 'red', 'orange', 'yellow', 'green','green'])
```
